### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
tiendas = [tienda, tienda2, tienda3, tienda4]
```

1. Análisis de facturación



```python
l_ventas_tienda = []
for i in range(4):

  ventas_tienda = tiendas[i]["Precio"].sum() - tiendas[i]["Costo de envío"].sum()
  print(f"PRECIO SUMA TIENDA {i+1}: {ventas_tienda}")
  l_ventas_tienda.append(ventas_tienda)
```

# 2. Ventas por categoría

```python
l_cantidad_cat = []
for i in range(4):

  cantidad_por_categoria = tiendas[i] ["Categoría del Producto"].value_counts()
  print(f"TIENDA {i+1} \n{cantidad_por_categoria}\n")
  l_cantidad_cat.append(cantidad_por_categoria)
```

# 3. Calificación promedio de la tienda


```python
l_calificacion = []
for i in range(4):
  calificacion = round(tiendas[i]["Calificación"].mean(),2)

  print(f"Promedio de calificación tienda {i+1}: ",calificacion)
  l_calificacion.append(calificacion)
```

# 4. Productos más y menos vendidos

```python
def max_y_min(producto):
  max = producto.max()
  min = producto.min()

  max_list = []
  min_list = []
  for i in producto.index:
    if producto[i] == max:
        max_list.append(i)

    if producto[i] == min:
        min_list.append(i)

  print(f"Productos más vendidos: {max_list}")
  print(f"Productos menos vendidos: {min_list} \n")

for i in range(4):
  producto = tiendas[i]["Producto"].value_counts()
  max_y_min(producto)
```

# 5. Envío promedio por tienda

```python
for i in range(4):
  valor_envio_t1 = tiendas[i]["Costo de envío"].mean()

  print(f"Valor promedio de envio tienda {i+1}: ", round(valor_envio_t1,2))
```

# Graficas

1. Ventas por tienda

```python
plt.bar(["tienda1","tienda2","tienda3","tienda4"],[l_ventas_tienda[0],l_ventas_tienda[1],l_ventas_tienda[2],l_ventas_tienda[3]], color = "red")
plt.title("Ventas totales por tienda")
plt.xlabel("Tienda")
plt.ylabel("Ingresos")
plt.show()
```

```python

plt.pie(l_cantidad_cat[0], labels=l_cantidad_cat[0].index, autopct='%1.1f%%')
plt.title ("Tienda1")
plt.show()
plt.pie(l_cantidad_cat[1], labels=l_cantidad_cat[0].index, autopct='%1.1f%%')
plt.title ("Tienda2")
plt.show()
plt.pie(l_cantidad_cat[2], labels=l_cantidad_cat[0].index, autopct='%1.1f%%')
plt.title ("Tienda3")
plt.show()
plt.pie(l_cantidad_cat[3], labels=l_cantidad_cat[0].index, autopct='%1.1f%%')
plt.title ("Tienda4")
plt.show()
```
